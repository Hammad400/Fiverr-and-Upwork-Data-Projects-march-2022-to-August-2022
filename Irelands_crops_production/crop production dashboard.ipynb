#### importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
pd.options.mode.chained_assignment = None  # default='warn'
```

#### importing dataset

```python
df=pd.read_csv('crop_production_modified.csv')
df.head()
```

#### Heat map of correalation

```python
# Finds correlation between Independent and dependent attributes
plt.figure(figsize = (10,10))
sns.heatmap(df.corr(), annot = True, cmap = "RdYlGn")
plt.title('Heatmap of correlation')
plt.show()
```

```python
df.columns
```

this heatmap tells us the relation between features so here i am intrested in the relation of features with crop yield per hectare in tonnes <br>
positive value of correlation shows positive relation and negative value of correlation shows negative relation.<br>
Year and Area_under_Crops have negative relation with Crop Yield per Hectare in Tonnes
crop production has positive value with crop yeild per hectare


#### analysing year

This graph shows the relationship betwween Years and crop yield per hectare using graph

```python
df.groupby('Year')['Crop_Yield_per_Hectare_in_Tonnes'].mean().plot(color='g')
plt.title('Change in Crop production per hecatre with respect to Year ')
plt.show()
```

result:After the year of 2000 A.D a great fall in production per hectare is occured so there is no positive relation between years and crop yield.

#### Graph of crop production per hectare

```python
df.groupby('Year')['crop_production_in_Tonnes'].mean().plot(color='g')
plt.title('Change in Crop production with respect to Year ')
plt.show()
```

this graph shows relation between years and crop production value.As in 2005 the crop production is at top. BUT There is also falling of crop production is observed in last year

#### Graph of Area under crops with respect to Year

Relation of year and crop production is drawn here.

```python
df.groupby('Year')['Area_under_Crops_Hectares'].mean().plot(color='g')
plt.title('Change in area under crops with respect to Year ')
plt.show()
```

here up and down grapgh tells that area is also increased and decreased with respect to Year

### analysing type of crop

#### Bar chart of Type of crops and Crop yeild per hectare 

```python
df.groupby('Type of Crop')['Crop_Yield_per_Hectare_in_Tonnes'].mean().plot(kind='barh',color='b')
plt.title('Relation of type of crops and Crop Yield per Hectare value using bar chart ')
plt.show()
```

this bar chart shows that Fodder beat,Sugar beet and Turnips are tpyes of crops with high yeild per hectare.

#### Bar chart of Type of crops and Crop crop_production

```python
df.groupby('Type of Crop')['crop_production_in_Tonnes'].mean().plot(kind='barh',color='b')
plt.title('Relation of type of crop and Crop Yield using bar chart ')
plt.show()
```

this bar chart shows that Total Wheat,oats and barley are tpyes of crops with high production

#### Note:<br>
It is difference between production or yield per hectare and production.<br>
production is simply mass in tonnes which is not good unit for production but in yield per hectare there is yield is compared with area so if high production with less area then there will be high value of yield per hectare 

#### Bar chart of Type of crops and Area under crops

```python
df.groupby('Type of Crop')['Area_under_Crops_Hectares'].mean().plot(kind='barh',color='b')
plt.title('Relation of type of crops and Area under crops using bar chart ')
plt.show()
```

this bar chart shows that Total Wheat,oats and barley are tpyes of crops with area under crops

So the answer of question about why Wheat,oats and barley gives high production is clear as more area is used so it gives high production
